## Initial visualization of Storms
- Using data from 2017

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
locations = '../StormEvents_locations-ftp_v1.0_d2017_c20190516.csv'
events = '../StormEvents_details-ftp_v1.0_d2017_c20190516.csv'
```

```python
locations_df = pd.read_csv(locations)
```

```python
locations_df.head()
```

```python
events_df = pd.read_csv(events)
```

```python
df = pd.DataFrame()
df['start_time'] = pd.to_datetime(events_df['BEGIN_DATE_TIME'])
df['end_time'] = pd.to_datetime(events_df['END_DATE_TIME'])
df['event_type'] = events_df['EVENT_TYPE']
df['id'] = events_df['EVENT_ID']

df['begin_lat'] = events_df['BEGIN_LAT']
df['begin_lon'] = events_df['BEGIN_LON']
df['end_lat'] = events_df['END_LAT']
df['end_lon'] = events_df['END_LON']
df['damage_property_in_k'] = events_df['DAMAGE_PROPERTY'].str[:-1].astype(float)
```

```python
df['episode_id'] = events_df['EPISODE_ID']
```

```python
combined_df = pd.merge(df, locations_df, left_on=['episode_id', 'id'], right_on=['EPISODE_ID', 'EVENT_ID'])
```
